```python
#open a file, read values to two arrays.  Plot arrays.
import matplotlib.pyplot as plt
import numpy as np   

x_array = np.zeros(0)
y_array = np.zeros(0)
with open("data001.txt", "r") as f:
    for line in f.readlines():
        res=line.split(",")
        a=float(res[0])
        b=float(res[1])
        #print(a,b)
        x_array = np.append(x_array,a)
        y_array = np.append(y_array,b)
        
        
x_array_2 = np.zeros(0)
y_array_2 = np.zeros(0)
with open("data002.txt", "r") as f:
    for line in f.readlines():
        res=line.split(",")
        a=float(res[0])
        b=float(res[1])
        #print(a,b)
        x_array_2 = np.append(x_array_2,a)
        y_array_2 = np.append(y_array_2,b)
        
        
x_array_3 = np.zeros(0)
y_array_3 = np.zeros(0)
with open("data003.txt", "r") as f:
    for line in f.readlines():
        res=line.split(",")
        a=float(res[0])
        b=float(res[1])
        #print(a,b)
        x_array_3 = np.append(x_array_3,a)
        y_array_3 = np.append(y_array_3,b)

fig = plt.figure(figsize=(12, 8))
plt.title('Laser spot height vs microstep number', fontsize=30)
plt.xlabel('Microstep (32)', fontsize=20)
plt.ylabel('Laser spot height, (mm)', fontsize=20)
plt.plot(x_array, y_array, '+r', markersize=8)

plt.plot(x_array_2, y_array_2, 'oy')

flip=np.flip(y_array_3)
plt.plot(x_array_3, flip, 'ob', markersize=8)

plt.show

plt.savefig('sample.svg')
```
