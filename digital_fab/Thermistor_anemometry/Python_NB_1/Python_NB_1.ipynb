```python
import matplotlib.pyplot as plt
import numpy as np;
import serial
ser = serial.Serial('/dev/tty.usbserial-FTFMJ5QQ')  # open serial port
print(ser.name)         # check which port was really used
#ser.flush()
x=arange(10)
y= np.zeros(10)
for i in range (0,10): y[i] = ord(ser.read() )         # read ten bytes
ser.close()             # close port
plt.plot(x, y, 'ro')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('serial data')
plt.show
```
