```python
import matplotlib.pyplot as plt
import numpy as np   

x_array = np.zeros(0)
y_array = np.zeros(0)
with open("data.txt", "r") as f:
    for line in f.readlines():
        res=line.split(",")
        a=float(res[0])
        b=float(res[1])
        #print(a,b)
        x_array = np.append(x_array,a)
        y_array = np.append(y_array,b)
#print(x_array)

plt.plot(x_array, y_array)
```
