```python
import matplotlib.pyplot as plt
import numpy as np   

x_array = np.zeros(0)
mag_array = np.zeros(0)
acc_array = np.zeros(0)

with open("./data10.txt", "r") as f:
        print(f.readline())
        for line in f.readlines():
            res=line.split(",")
            a=float(res[0])
            b=float(res[1])
            c=float(res[2])
        
            x_array = np.append(x_array,a)
            mag_array = np.append(mag_array,b)
            acc_array = np.append(acc_array,c)

f.close()

indx_1 = np.argmin(mag_array[0:100])
next = np.argmin(mag_array[indx_1+10:indx_1+150])
indx_2 = next + indx_1 +10
T = indx_2-indx_1

sin_array = np.sin((2*np.pi/T)*(x_array - indx_1))
cos_array = np.cos((2*np.pi/T)*(x_array - indx_1))

plt.plot(x_array, 25*sin_array)
plt.plot(x_array, 25*cos_array)



in_phs = sin_array * (acc_array+255)
out_phs = cos_array * (acc_array+255)

in_amp = sum(in_phs)/300
out_amp = sum(out_phs)/300


plt.plot(x_array, mag_array/250, "or", markersize=4)
plt.plot(x_array, 255+acc_array, "oy", markersize=4)

in_amp, out_amp, np.sqrt(in_amp**2 + out_amp**2)
```
